### To read a csv file using spark by creating dataframe

```python
%python
df1=spark.read.csv("dbfs:/Volumes/workspace/default/storage_device/two.csv", header=True, inferSchema=True)
display(df1.limit(5))
```

### To read a excel file using spark by creating dataframe

```python
%python
df_excel=spark.read.format("excel").option("header","true").option("inferschema", "True").load("/Volumes/workspace/default/mydrive/Empl2.xlsx")
display(df_excel.limit(5))
```

### Saving excel file as table 

```python
%python
df_excel.write.saveAsTable("emp_details")
```

### Listing the created table using spark sql

```python
%sql
select *from emp_details;
```

### Downloading sample CSV data from a webportal using python API call and storing it in a volume

```python
%python
import requests
url="https://www.stats.govt.nz/assets/Uploads/Business-operations-survey/Business-operations-survey-2022/Download-data/business-operations-survey-2022-information-and-communications-technology.csv"
response=requests.get(url)
csv_data=response.text
dbutils.fs.put("/Volumes/workspace/default/storage_device/business_data.csv",csv_data, overwrite=True)
```

### Reading the sample CSV

```python
%python
df2=spark.read.csv("/Volumes/workspace/default/storage_device/business_data.csv",header=True, inferSchema=True)
display(df2.limit(5))
df2.printSchema()
       
```
